```python
# credits: https://gist.github.com/ahmetalsan/06596e3f2ea3182e185a
import re
import math
from collections import Counter

def get_cosine(vec1, vec2):
    intersection = set(vec1.keys()) & set(vec2.keys())
    numerator = sum([vec1[x] * vec2[x] for x in intersection])

    sum1 = sum([vec1[x]**2 for x in vec1.keys()])
    sum2 = sum([vec2[x]**2 for x in vec2.keys()])
    denominator = math.sqrt(sum1) * math.sqrt(sum2)

    if not denominator:
        return 0.0
    else:
        return float(numerator) / denominator

def text_to_vector(text):
    word = re.compile(r'\w+')
    words = word.findall(text)
    return Counter(words)


def cosine(content_a, content_b):
    text1 = content_a
    text2 = content_b

    vector1 = text_to_vector(text1)
    vector2 = text_to_vector(text2)

    cosine_result = get_cosine(vector1, vector2)
    return cosine_result
```

```python
''' Readme:

The mXm dataset comes in 2 text files, train and test.

To simplify its usage, we also provide a SQLite database with the
same data. The code to recreate this database is:
   mxm_dataset_to_db.py
More details on the database:
   - it contains two tables, 'words' and 'lyrics'
   - table 'words' has one column: 'word'. Words are entered according
     to popularity, check their ROWID if you want to check their position.
     ROWID is an implicit column in SQLite, it starts at 1.
   - table 'lyrics' contains 5 columns, see below
   - column 'track_id' -> as usual, track id from the MSD
   - column 'mxm_tid' -> track ID from musiXmatch
   - column 'word' -> a word that is also in the 'words' table
   - column 'count' -> word count for the word
   - column 'is_test' -> 0 if this example is from the train set, 1 if test

If you want to know exactly how we created the bag-of-wirds, look at:
  lyrics_to_bow.py
Note that it requires the following Python package:
  http://pypi.python.org/pypi/stemming/1.0


# musiXmatch dataset - list of matches
#     matches provided by musiXmatch based on artist names
#     and song titles from the Million Song Dataset.
# MORE INFORMATION:
#     http://labrosa.ee.columbia.edu/millionsong/musixmatch
# FORMAT:
#     #    -> comment, ignore
#     tid|artist name|title|mxm tid|artist_name|title
#        tid          -> Million Song Dataset track ID
#        artist name  -> artist name in the MSD
#        title        -> title in the MSD
#        mxm tid      -> musiXmatch track ID
#        artist name  -> artist name for mXm
#        title        -> title for mXm
#        |            -> actual separator: <SEP>
# QUESTIONS / COMMENTS
#     tb2332@columbia.edu
# Enjoy!
'''
with open('mxm_779k_matches.txt') as file:
    mxm_song_dataset = [line.rstrip().split('<SEP>') for line in file]
```

```python
# v2
def find_artist_songs(artist, song_list):
  artist_songs = []
  for song in song_list:
    musixmatch_artist = unidecode(song[4].lower())
    if musixmatch_artist == artist: # or cosine(artist, musixmatch_artist) >= 0.95:
      artist_songs.append(song)
  return artist_songs  
    
found_songs = []
cosines = []
for i in range(len(songs)):  
  max_cosine = 0
  best_match = None
  song_name = unidecode(songs[i][2].lower())
  song_author = unidecode(songs[i][3].lower())
  # Filter cases in which the song author's name is already on the song name.
  if song_author in song_name:
    song_name = song_name.replace(song_author, '')
  # Find whether there are songs in the MusixMatch database 
  # that are associated with this artist.
  artist_songs = find_artist_songs(song_author, mxm_song_dataset)
  # Whithin an artist song list search for the most similar song title in MusixMatch dataset.
  for song in artist_songs:
    musixmatch_song_name = unidecode(song[5].lower())
    musixmatch_song_author = unidecode(song[4].lower())
    # Remove artist name from song title if present.
    if musixmatch_song_author in musixmatch_song_name:
      musixmatch_song_name = musixmatch_song_name.replace(musixmatch_song_author, '')
      
    test_cosine = cosine(song_name, musixmatch_song_name)
    if test_cosine >= max_cosine:
      max_cosine = test_cosine
      best_match = song
  cosines.append(max_cosine)
  found_songs.append((songs[i], best_match))
  print_progress(count=i, total=len(songs)-1)

# Filter matches with accuracy greater than 65%  
matched = []
for i in range(len(found_songs)):
  if cosines[i] >= 0.65:
    matched.append([found_songs[i][0],found_songs[i][1]])
```

```python
# Helper fn.
def print_progress(count, total):
    # Percentage completion.
    pct_complete = float(count) / total
    # Status-message.
    # Note the \r which means the line should overwrite itself.
    msg = "\r- Progress: {0:.1%}".format(pct_complete)
    # Print it.
    sys.stdout.write(msg)
    sys.stdout.flush()

def create_connection(db_file):
    """ create a database connection to the SQLite database
        specified by the db_file
    :param db_file: database file
    :return: Connection object or None
    """
    conn = None
    try:
        conn = sqlite3.connect(db_file)
    except Error as e:
        print(e)

    return conn

def select_words_by_music_id(connection, mxm_id):
  cursor = connection.cursor()
  cursor.execute("SELECT word, count FROM lyrics WHERE track_id=?", (mxm_id,))
  rows = cursor.fetchall()
  return rows

# Find the words and occurrences on mxm lyrics dataset and join with the matched songs. 
connection = create_connection('mxm_dataset.db')
for match in matched:
  mxm_id = match[1][0]
  match.append(select_words_by_music_id(connection, mxm_id))
  
# Filter brazilian ones from lyrics for word occurrences in portuguese dictionary with accuracy greater than 51%. 
brazilian_songs = []
for match in matched:
  lyrics = match[2]
  score = 0
  if len(lyrics) > 0:
    score = sum([word[0] in word_dictionary for word in lyrics])
    if (score / len(lyrics)) >= 0.51:
      brazilian_songs.append(match)
```
